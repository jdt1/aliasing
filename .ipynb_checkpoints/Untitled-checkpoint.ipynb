```python
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
import numpy as np
from math import pi
```

```python
f_sample = 100     # [Hz] / [fps]
f_signal = 1234    # [Hz]
```

```python
N = 1000000        # [-] number of points to calculate
```

```python
p_sample = 1/f_sample  # [s]
p_signal = 1/f_signal  # [s]
```

```python
t = np.linspace(0,1,N)
```

```python
signal = np.sin(t*f_signal*2*pi)
sample_times = np.linspace(0,1,f_sample)
sampled_signal = np.sin(2*pi*f_signal*sample_times)
```

```python
# plt.plot(sampled_signal)
plt.plot(sample_times, sampled_signal)
plt.xlim(0,sample_times[5])
```

```python
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

yf = rfft(sampled_signal)
xf = rfftfreq(f_sample, 1 / f_sample)

plt.plot(xf, np.abs(yf))
plt.xlim(0,f_sample)
plt.show()
```
